```python
data=pd.read_csv(r'C:\Users\vaibhav mathur\Desktop\Saloni\Titanic_ISTE ML.csv')
encoder=ce.OrdinalEncoder(cols=['Sex','Embarked'],return_df=True,mapping=[{'col':'Sex','mapping':{'male':0,'female':1}}, 
                                                                          {'col':'Embarked','mapping':{'S':0,'C':1,'Q':2}}])
data_transformed=encoder.fit_transform(data)
#data_transformed['Embarked'].fillna(3,inplace=True)
data_transformed.head()
```

```python
correlation1=data_transformed['Survived'].corr(data_transformed['Pclass'])
correlation2=data_transformed['Survived'].corr(data_transformed['Sex'])

correlation3=data_transformed['Survived'].corr(data_transformed['Age'])
correlation4=data_transformed['Survived'].corr(data_transformed['SibSp'])
correlation5=data_transformed['Survived'].corr(data_transformed['Fare'])

correlation6=data_transformed['Survived'].corr(data_transformed['Embarked'])
print(correlation1, correlation2, correlation3, correlation4, correlation5, correlation6, sep='\n')
```

```python
# chi square method
from scipy.stats import chi2_contingency
```

```python
crosstabresult1=pd.crosstab(index=data['Survived'],columns=data['Pclass'])
ChiSqResult1=chi2_contingency(crosstabresult1)
print('The P-Value of the ChiSq Test is:', ChiSqResult1[1])

crosstabresult2=pd.crosstab(index=data['Survived'],columns=data['Sex'])
ChiSqResult2=chi2_contingency(crosstabresult2)
print('The P-Value of the ChiSq Test is:', ChiSqResult2[1])

crosstabresult3=pd.crosstab(index=data['Survived'],columns=data['Age'])
ChiSqResult3=chi2_contingency(crosstabresult3)
print('The P-Value of the ChiSq Test is:', ChiSqResult3[1])

crosstabresult4=pd.crosstab(index=data['Survived'],columns=data['SibSp'])
ChiSqResult4=chi2_contingency(crosstabresult4)
print('The P-Value of the ChiSq Test is:', ChiSqResult4[1])

crosstabresult5=pd.crosstab(index=data['Survived'],columns=data['Fare'])
ChiSqResult5=chi2_contingency(crosstabresult5)
print('The P-Value of the ChiSq Test is:', ChiSqResult5[1])

crosstabresult6=pd.crosstab(index=data['Survived'],columns=data['Embarked'])
ChiSqResult6=chi2_contingency(crosstabresult6)
print('The P-Value of the ChiSq Test is:', ChiSqResult6[1])
```

# Q2

The Survived Column data has only two missing values so those can be changed with the mode value, which will not cause much
change in the actual data or the output

```python
data_transformed['Survived'].fillna(0,inplace=True) 
```

The Pclass column data has only one missing value which can be replaced with the mode value 3.Since the probablity of that data to be 3 is higher

```python
data_transformed['Pclass'].fillna(3.0,inplace=True)
```

The Sex column has 10 missing values which could be replaced with the mode value(0 that is male).Since mean,median cannot be applied on this mode will be the best to use

```python
data_transformed['Sex'].fillna(0,inplace=True)
```

The missing values in the age column can be replaced with 29 since mean of the data is 29.67 and the median is 28 so taking a value between the both

```python
data_transformed['Age'].fillna(29,inplace=True)
```

Since there is only one missing value in the Sibsp column it can be replaced dwith 0 since both mode and median are 0

```python
data_transformed['SibSp'].fillna(0,inplace=True)
```

The two null values in Parch column can be replace with 0 as both mode and median are eual to 0

```python
data_transformed['Parch'].fillna(0,inplace=True)
```
